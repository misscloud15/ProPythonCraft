**PYTHON DATA TYPES**


int (Integer)

Stores whole numbers without decimals.
Properties: immutable, supports arithmetic operations, unlimited size (depends on memory).
Example: x = 10

```python
x = 1
y = 2
z = x+y
print(z)
print(type(x))
```


float

Stores decimal (real) numbers.
Properties: immutable, supports arithmetic operations, less precise than integers due to floating-point representation.
Example: pi = 3.14

```python
a = 2.345
print(type(a))
```

complex

Stores complex numbers with real and imaginary parts.
Properties: immutable, used in scientific and mathematical calculations.
Example: z = 2 + 3j

```python
z = 2 + 3j
print(type(z))
```

bool (Boolean)

Stores logical values.
Properties: immutable, has only two values: True and False, often used in conditions.
Example: is_active = True

```python
is_active = True
print(type(is_active))
```

list

Stores a collection of items.
Properties: mutable, ordered, allows duplicate values, can store mixed data types.
Example: nums = [1, 2, 3, "a"]

```python
a = [1,3,5,7]
print(type(a))
```

tuple

Stores a collection of items similar to a list.
Properties: immutable, ordered, allows duplicates, faster than lists.
Example: point = (10, 20)

```python
t = (2,4,6,8,10)
print(type(t))
```

set

Stores unique items only.
Properties: mutable, unordered, does not allow duplicates, supports set operations (union, intersection).
Example: unique = {1, 2, 3}

```python
s = {1,2,3,5}
print(type(s))
```
