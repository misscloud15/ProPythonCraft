In Python, literals are fixed values written directly in the code. They represent constant data that does not change while the program runs.

INTEGER LITERALS

```python
a = 1008
b = 12_4578_6  # for easier reading b = _344 , b = 34_  gives error
print(a,b)
```

```python
b = _344
```

```python
34_
```

FLOAT LITERALS

```python
a = 4540.56
b = 29_46.56
print(a,b)
```

```python
c = 244_._46
```

BOOL LITERALS

```python
A = True
B = False
print(A,B)
```

COMPLEX LITERALS

```python
B = 4_5+5_6j
print(B)
```

STRING LITERALS

```python
first ='Megha'
second ="Megha"
third ='''Megha'''
print(first,second,third)
```
